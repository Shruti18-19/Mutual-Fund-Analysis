```python
plt.xlabel("Returns")
plt.ylabel("Frequency")
plt.title("Distribution of returns")
plt.hist(new_df['returns_1yr'],color='blue',label="1 year")
plt.hist(new_df['returns_3yr'],color='red',label="3 years")
plt.hist(new_df['returns_5yr'],color="yellow",label="5 years")
plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()
```

```python
return1 = [maximum1,minimum1,average1]
return3 = [maximum3,minimum3,average3]
return5 = [maximum5,minimum5,average5]
l1 = ["Maximum","Minimum","Average"]
plt.plot(l1,return1,color='red',label="1 year")
plt.plot(l1,return3,color='blue',label="3 years")
plt.plot(l1,return5,color='yellow',label="5 years")
plt.title("Returns rate")
plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()
plt.show()
```

```python
numeric_cols = ["sortino", "alpha", "sd", "beta", "sharpe"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

heatmap_data = df[["expense_ratio", "fund_size_cr", "fund_age_yr", "sortino", 
                    "alpha", "beta", "sharpe", "sd", "risk_level", "rating", 
                    "returns_1yr", "returns_3yr", "returns_5yr"]]

heatmap_data = heatmap_data.dropna()

corr_matrix = heatmap_data.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Correlation Heatmap of Mutual Funds Attributes")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["fund_size_cr"], y=df["returns_1yr"], hue=df["risk_level"], palette="coolwarm", size=df["rating"], sizes=(20, 200))
plt.xlabel("Fund Size (in Crores)")
plt.ylabel("1-Year Returns (%)")
plt.title("Scatter Plot: 1-Year Returns vs Fund Size")
plt.legend(title="Risk Level", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["risk_level"], y=df["returns_1yr"], palette="coolwarm")
plt.xlabel("Risk Level")
plt.ylabel("1-Year Returns (%)")
plt.title("Box Plot: 1-Year Returns Distribution by Risk Level")
plt.show()
```
